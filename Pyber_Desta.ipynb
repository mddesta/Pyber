```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Load and read the data files 
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city_data=pd.read_csv(city_data_to_load)
ride_data=pd.read_csv(ride_data_to_load)

# Combine the data into a single dataset
pyber_data=pd.DataFrame(pd.merge(ride_data,city_data,on="city",how="left"))

# Display the data table for preview
pyber_data.head()
```

## Bubble Plot of Ride Sharing Data

```python
#Urban Scatter plot
urban_city=pyber_data[pyber_data["type"]=="Urban"].groupby(["city"])
urbanx_axis=urban_city["ride_id"].count()
urbany_axis=urban_city["fare"].mean()
urban_drivercount=urban_city["driver_count"].mean()
Urban=plt.scatter(urbanx_axis,urbany_axis, marker="o",facecolors="coral", edgecolors="black",alpha=0.7, linewidth=1, s=urban_drivercount*10,label="Urban")

#Suburban Scatter plot
suburban_city=pyber_data[pyber_data["type"]=="Suburban"].groupby(["city"])
suburbanx_axis=suburban_city["ride_id"].count()
suburbany_axis=suburban_city["fare"].mean()
suburban_drivercount=suburban_city["driver_count"].mean()
Suburban=plt.scatter(suburbanx_axis,suburbany_axis, marker="o",facecolors="skyblue", edgecolors="black",alpha=0.7, linewidth=1, s=suburban_drivercount*10,label="Suburban")

#Rural Scatter plot
rural_city=pyber_data[pyber_data["type"]=="Rural"].groupby(["city"])
ruralx_axis=rural_city["ride_id"].count()
ruraly_axis=rural_city["fare"].mean()
rural_drivercount=rural_city["driver_count"].mean()

Rural=plt.scatter(ruralx_axis,ruraly_axis, marker="o",facecolors="gold", edgecolors="black",alpha=0.7, linewidth=1, s=rural_drivercount*10,label="Rural")

#add grid lines
plt.grid()
#add legend
legend=plt.legend(handles=[Urban, Suburban, Rural],fontsize=8,loc='upper right')

#add title and axis names
plt.title("Pyber Ride Sharing Data (2016)")
plt.xlabel("Total Numbers of Rides (Per City)")
plt.ylabel("Average Fare ($))")

# Adjust the legend 
legend.legendHandles[0]._sizes = [30]
legend.legendHandles[1]._sizes = [30]
legend.legendHandles[2]._sizes = [30]

#add note
plt.text(42,38,'Note:') 
plt.text(42,36,'Circle size correlates with driver count per city.')
#Save Figure
plt.savefig("Output/Ride Sharing Data.png")

# Show plot
plt.show()
```

## Total Fares by City Type

```python
# Calculate Type Percents
fares=pyber_data.groupby(["type"])
total_fares=pd.DataFrame(fares["fare"].sum())
total_fares
```

```python
# Build Pie Chart
explode = (0, 0, 0.1)
colors="gold","skyblue", "lightcoral"
plt.pie(total_fares["fare"], explode=explode, labels=total_fares.index, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=150)
plt.axis("equal")
#add Title
plt.title('% of Total Fares by City Type')

#Save Figure
plt.savefig("Output/Total Fares.png")

# Show Figure
plt.show()
```

## Total Rides by City Type

```python
# Calculate Ride Percents
total_rides=pd.DataFrame(fares["fare"].count())
total_rides
```

```python
# Build Pie Chart
explode = (0, 0, 0.1,)
colors="gold","skyblue", "lightcoral"
plt.pie(total_rides["fare"], explode=explode, labels=total_rides.index, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=150)
plt.axis("equal")

#add Title
plt.title('% of Total Rides by City Type')

# Save Figure
plt.savefig("Output/Total Rides.png")

plt.show()
```

## Total Drivers by City Type

```python
# Calculate Driver Percents
drivers=city_data.groupby(["type"])
total_drivers=pd.DataFrame(drivers["driver_count"].sum())
total_drivers
```

```python
# Build Pie Charts
explode = (0, 0, 0.15,)
colors="gold", "skyblue", "lightcoral"
plt.pie(total_drivers["driver_count"], explode=explode, labels=total_drivers.index, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=150)
plt.axis("equal")

#add Title
plt.title("% of Total Drivers by City Type")

#Save Figure
plt.savefig("Output/Total Drivers.png")

# Show Figure
plt.show()
```
